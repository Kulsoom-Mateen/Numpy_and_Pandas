# Essential functionality

### Reindexing on series

```python
obj = pd.Series([4.5,7.2,-5.3,3.6] , index=['d','b','a','c'])
```

```python
obj
```

```python
obj1 = obj.reindex(['a','b','c','d','e'])
obj1
```

```python
obj2 = pd.Series(['blue','purple','yellow'],index=[0,2,4])  
```

```python
obj2.reindex(range(6))    #undefined indexes will get NaN
```

```python
obj2.reindex(range(6),method='ffill')   #ffill=forward fill    To remove NaN
```

```python
obj2.reindex(range(6),method='bfill')   #bfill=backward fill    To remove NaN  5 remain NaN bcz it doesn't get its value from anyone on the back
```
