```python
import numpy as np
import pandas as pd
```

# Merge or join operations

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                     'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                     'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)          #by default how='inner' so it will pick only those values which are common in both dataFrame
```

```python
df1e = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'key1':[1,2,3,4,5,6,7],
                     'data1': range(7)})
df2e = pd.DataFrame({'key': ['a', 'b', 'd','b'],'key1':[1,2,3,4],
                     'data2': range(4)})
#here pair of key will be searched in one frame whether it is also available in second frame.if present in both show their result
```

```python
df1e
```

```python
df2e
```

```python
pd.merge(df1e, df2e)
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                     'data1': range(7)})
df3
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                  'data2': range(3)})
df4         
```

```python
#columns name of df3 and df4 are different so we have to tell it
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
#left dataframe i.e df3 me se 'lkey' column ki basis p
#right datframe i.e df4 me se 'rkey' column ki basis p merge krna hai
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey' , how='outer')
#it perform union
#include all keys whether they are similar(forming pair) or not
```

```python
pd.merge(df1, df2, on='key', how='left')
#jo common nhi hain un me se left wali value lega
```

```python
pd.merge(df1, df2, on='key', how='right')
```

# Join

```python
# Merge performs on columns basis
# Join performs on index basis by default but we can change it to columns
df3.join(df4)
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                      index=['a', 'c', 'e'],
                      columns=['Ohio', 'Nevada'])
left2
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['Missouri', 'Alabama'])
right2
```

```python
left2.join(right2)
#here it remove 2 keys
```

```python
left2.join(right2 , how='outer')
```

```python
left2.join(right2)     #here how='left'      Show all of left(left2) i.e a,c,e
```

```python
left2.join(right2 , how='right')       #show all of right i.e b,c,d,e
```

```python
#in 2 datframes 2 columns are same so on which column basis we want to merge
#there are 2 keys in both df1e and df2e i.e. 'key' and 'key1' so we tell it to perform merge on the basis of 'key'
#if we don't put suffix then it will by default show key1_x and key1_right
pd.merge(df1e,df2e, on='key' , suffixes=('_left','_right'))
```

```python
df1d = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                  columns=['one', 'two'])
df1d
```

```python
df1p = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                     columns=['one', 'two'])
df1p
```

```python
df2p = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'],
                     columns=['three', 'four'])
df2p
```

```python
pd.concat([df1p, df2p], axis=1, keys=['level1', 'level2'] , sort='True')
```
