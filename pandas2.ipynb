## Correlation and covariance

```python
#how closely two dataframes or two series are related to each other 
```

```python
import pandas_datareader.data as web
```

```python
all_data = {ticker: web.get_data_yahoo(ticker)    #all data is dictionary and ticker is key
 for ticker in ['AAPL', 'IBM', 'MSFT', 'GOOG']}   #ticker is a variable and every time value will change
all_data
```

```python
all_data["AAPL"]
```

```python
price = pd.DataFrame({ticker: data['Adj Close']
 for ticker, data in all_data.items()})
```

```python
price
```

```python
all_data.keys()
```

```python
volume = pd.DataFrame({ticker: data['Volume']
 for ticker, data in all_data.items()})
volume
```

```python
#pct_change  == percentage change
returns = price.pct_change()
returns
```

```python
pd.options.display.max_rows=10
```

```python
returns["MSFT"].corr(returns["IBM"])
```

```python
returns['MSFT'].cov(returns['IBM'])
```

```python
returns.corr()  #over all correlation 
```

```python
returns.cov()  #over all covariance 
```

```python
returns.corrwith(returns.IBM)
```
